```python
#Importing of data
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
#Checking first few lines of data (default is 5)
df.head()
```

```python
#Checking all the columns and its data type
df.info()
```

```python
#Summary Statistics
df.describe() #By default it will give summary stats for only numerical columns
#If you want to include everything in summary stats use: df.describe(include='all')
```

```python
#Checking for any null values
df.isnull().sum()
```

```python
#Filling in the null values using the median because it is robust to the outliers unlike mean.
#For this case we will be grouping the review rating as per the product category in order to avoid bias
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
#Checking for null values again
df.isnull().sum()
```

```python
#Replacing your column name to snake casing.
#Snake casing it transforming you column name in all lowercase with underscores.
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
#Checking column names
df.columns
```

```python
#Creating a new column called age_group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels) #qcut splits the ages into four equal-sized groups 
                                                           #based on the data distribution and assigns the labels we defined.
```

```python
#Checking specific column data
df[['age', 'age_group']].head(10) #Specifing the number of data lines that you want to check
```

```python
#Creating a column named purchase_frequency_days
#The data in frequency of purchase column is in text form, we will be convertong it into numerical form
#Creating a dictionary called frequency mapping that maps each textual frequency to the equivalent no. of days.

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
#Checking if the data of both the columns are same or not
#If true then yes and if false then no.

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#Dropping the column named promo_code_used

df = df.drop('promo_code_used', axis=1)
```
